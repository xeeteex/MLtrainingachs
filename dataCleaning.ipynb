# Group By and Aggregating

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/AI ML session/Flavors.csv")
df
```

```python
group_by_frame = df.groupby("Base Flavor")
```

```python
group_by_frame.mean(numeric_only= True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.sum()
```

```python
df.groupby(["Base Flavor", "Liked" ]).aggregate({ "Flavor Rating": ["mean", "max", "count", "sum"], "Texture Rating": ["mean", "max", "count", "sum"]} )
```

```python
desc = df.groupby(["Base Flavor", "Liked" ]).aggregate({ "Flavor Rating": ["mean", "max", "count", "sum"], "Texture Rating": ["mean", "max", "count", "sum"]} )
desc.describe()
```

#Data Merge, Join and Concatenate

```python
import pandas as pd
df1 = pd.read_csv("/content/drive/MyDrive/AI ML session/LOTR.csv")
df2 = pd.read_csv("/content/drive/MyDrive/AI ML session/LOTR 2.csv")
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

# Concatenate df1 and df2

```python
df = pd.concat([df1, df2])
```

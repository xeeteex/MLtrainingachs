## Why matplotlib and seaborn?
Both libraries are among the popular choice for visualzation in python.
Here's why :
- Matplotlib: Known for flexibility , Matplotlib allows detailed customization of visual elements, making it perfect for tailored plots
- Seaborn: Built on top of Matplolib , Seaborn simplifies visualization tasks by offering high level functions to create aesthetically pleasing statistical graphics

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y , label= "sine wave")
plt.title(" Sine wave Line plot")
plt.xlabel(" Value of x")
plt.ylabel(" Value of y")
plt.legend()
plt.show()
```

#Bar chart

```python
categories = ['A', 'B', 'C']
values = [40, 120, 80]
plt.bar(categories, values, color= ["Red", "Blue", "Green"])
plt.title(" Bar chart example")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

# Histogram and KDE plot


```python
#trial
weights = [2, 4, 3, 1, 5, 7, 2, 3, 3, 1, 4, 2]
plt.hist(weights)
plt.title("Weight change throughout the days")
plt.xlabel("days")
plt.ylabel("weights")
plt.show()
```

```python
data = np.random.randn((1000))
sns.histplot(data , kde = True, color = "green")
plt.title("Histogram with KDE")
plt.show()
```

# Heatmap

```python
matrix = np.random.rand(5,5)
sns.heatmap(matrix, annot = True, cmap= "coolwarm") #annot tyo each sec ma num plot gareko
plt.title("heatmap example")
plt.show()
```

#Pair Plot

```python
sns.set_style ("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue= "species", markers= ["o", "s", "D"])
plt.show()
```
